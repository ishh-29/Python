**Data Cleaning,Preprocessing And Handling Missing Values**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Detect Missing Values Of A Given DataFrame.Display True Or False.

```python
pd.set_option('display.max_rows',None)
#pd.set_option('display.max_columns',None)
df=pd.DataFrame({
'ord_no':[70001,np.nan,70002,70004,np.nan,70005,np.nan,70010,70003,70012,np.nan,70013],
'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45, 75.29,3045.6],
'ord_date': ['2012-10-05','2012-09-10',np.nan,'2012-08-17','2012-09-10','2012-07-27','2012-09-10','2012-10-10','2012-10-10','2012-06-27','2012-08-17','2012-04-25'],
'customer_id':[3002,3001,3001,3003,3002,3001,3001,3004,3003,3002,3001,3001],
'salesman_id':[5002,5003,5001,np.nan,5002,5001,5001,np.nan,5003,5002,5003,np.nan]
})
print(df)
print("\nMissing Values Of The Said DataFrame:")
print(df.isna())
```

2. Identify The Column(s) Of A Given DataFrame Which Have At Least One Missing Value.

```python
print(df.isna().any())
```

3. Drop The Rows Where At Least One Element Is Missing In A Given DataFrame.

```python
res=df.dropna()
print(res)
```

4. Calculate The Total Number Of Missing Values In A DataFrame.

```python
res=df.isna().sum().sum()
print(res)
```

5. Replace NaNs With A Single Constant Value In Specified Columns In A DataFrame.

```python
print("\nReplacing NaNs With A Single Constant Value:")
res=df['ord_no'].fillna(0,inplace=True)
print(res)
```

6. Replace NaNs With The Value From The Previous Row Or The Next Row In A Given DataFrame.

```python
print("\nReplacing NaNs With The Value From The Next Row:")
df['purch_amt'].fillna(method='pad',inplace=True)
print(df)
print("\nReplacing NaNs With The Value From The Previous Row:")
df['sale_amt'].fillna(method='bfill',inplace=True)
print(df)
```
