1. Add,Substract,Multiply And Divide Two Pandas Series

```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
from collections import Counter
```

```python
ds1=pd.Series([2,4,6,8,10])
ds2=pd.Series([1,3,5,7,9])
op=input("Enter Operator:")
match op:
  case "+":
    print(ds1+ds2)
  case "-":
    print(ds1-ds2)
  case "*":
    print(ds1*ds2)
  case "/":
    print(ds1/ds2)
  case _:
    print("Invalid Operator")
```

13. Get The Items Which Aren't Common Of Two Given Series.

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([2,4,6,8,10])
s11=pd.Series(np.union1d(s1,s2))
s22=pd.Series(np.intersect1d(s1,s2))
res=s11[~s11.isin(s22)]
print(res)
```

23. Get The Day Of Month,Day Of Year,Week Number And Day Of Week From A Given Series Of Date Strings.

```python
dates=pd.Series(['01 Jan 2015','10-02-2016',
                 '20180307','2014/05/06','2016-04-12','2019-04-06T11:20'])
print("Original Series:")
print(dates)
dates=dates.map(lambda x:parse(x))
print("Day Of Month:")
print(dates.dt.day.tolist())
print("Day Of Year:")
print(dates.dt.dayofyear.tolist())
print("Week Number:")
print(dates.dt.weekofyear.tolist())
print("Day Of Week:")
print(dates.dt.weekday_name.tolist())
```

26. Find The Positions Of The Values Neighboured By Smaller Values On Both Sides In A Given Series.

image.png

```python
nums=pd.Series([1,8,7,5,6,5,3,4,7,1])
print(nums)
print("\nPositions Of The Values Of Peaks:")
'''
Firstly Computing The Second Order Difference OF The 'nums' Series
Using The np.diff and np.sign Function Then Difference Of Consecutive
Elements Of The Resulting Sign Series
'''
temp=np.diff(np.sign(np.diff(nums)))
'''
Difference Is Given By out[i]=a[i+1]-a[i] Along
The Given Axis,Higher Differences Are Calculated By Using diff
Recursively.
Sign Function Returns -1 if x<0,0 if x==0, 1 if x>0.
nan Is Returned For NaN Inputs.
'''
res=np.where(temp==-2)[0]+1
'''
Finds The Positions In The Secod Order Difference Series
Where The Value Is Equal To -2.
where() Fucntions Returns A Tuple Containing The Indices Where
This Condition Is True.
'''
print(res)
```

27. Compute The Autocorrelations Of A Given Numeric Series.

**Autocorrelation***: The Correlation Of A Signal With A delayed Copy Of Itself As A Function Of Delay.

Its The Similarity Between Observations As A Function Of The Time Lag Between Them.

Also Known As Serial Correlation.

```python
nums=pd.Series(np.arange(15)+np.random.normal(1,10,15))
print(nums)
autoco=[nums.autocorr(i).round(2) for i in range(11)]
'''
Creates A Series Where Each Element Is The Sum Of A
Corresponding Element In A Range Of 15 Numbers And A Random Number Drawn
From A Normal Distribution With A Mean Of 1 And Standard
Deviation Of 10
'''
print("\nAutocorrelation Of The Said Series:")
print(autoco[1:])
```

16. Count City Wise Number Of People From A Given Data Set (city,name of the person)

```python
df=pd.DataFrame({'name':['A','B','C','D','E','F','G','H','I','J'],
                    'city':['1','2','1','1','1','2','2','3','3','2']})
g=df.groupby(["city"]).size().reset_index(name='Number Of People:')
'''
groupby() Groups The Rows Of df By The 'city' Column
size() Method Counts Number Of Rows In Each Group
reset_index() Method Resets The Index Of The Resulting DataFrame And
renames The Column To 'Number Of People:'
'''
print(g)
```
