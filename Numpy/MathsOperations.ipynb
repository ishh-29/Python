```python
import numpy as np
```

1. **To Compute Logarithm Of The Sum Of Exponentations Of The Inputs,Sum Of Exponentiatios Of The Input In Base-2**

np.logaddexp()->Performs The Computation Of log(exp(x1​)+exp(x2​))

np.logaddexp2()->Base-2 Logarithm Version Of np.logaddexp().Performs The Computation Of log2(2^x1+2^x2)

```python
l1=np.log(1e-50)
l2=np.log(2.5e-50)
print("Logarithm Of The Sum Of Exponentiations:")
print(np.logaddexp(l1,l2))
print(np.logaddexp2(l1,l2))
```

2. **To Get True Division Of The Element-Wise Array Inputs**

np.true_divide()->Performs True (floating-point) Division Element-wise Ensuring That The Result Is Always A Float,Even If The Inputs Are Integers

```python
x=np.arange(10)
print(np.true_divide(x,3))
```

3. **To Get The Largest Integer Smaller Or Equal To The Division Of The Inputs**

np.floor_divide()->Performs Element-wise Floor Division

```python
x=[1.,2.,3.,4.]
print(np.floor_divide(x,1.5))
```

4. **To Get The Powers Of An Array Values Elements-Wise**

```python
x=np.arange(7)
print(np.power(x,2))
```

5. **To Get Element-Wise Remainder Of Array Elements**

6. **To Multiply A 5x3 Matrix By A 3x2 Matrix And Create A Real Matrix Product**

```python
x=np.random.random((5,3))
y=np.random.random((3,2))
print(np.dot(x,y))
```

7. **To Multiply A Matrix By Another Matrix Of Complex Numbers And Create A New Matrix Of Complex Number**

```python
x=np.array([1+2j,7+8j])
y=np.array([5+6j,7+8j])
print(np.vdot(x,y))
```

8. **To Create An Inner Product Of Two Arrays**

```python
x=np.arange(24).reshape((2,3,4))
y=np.arange(4)
print(np.inner(x,y))
```

9. **To Generate Inner,Outer And Cross Products Of Matrices And Vectors**

```python
x=np.array([1,4,0],float)
y=np.array([2,2,1],float)
print(np.inner(x,y))
print(np.outer(x,y))
print(np.cross(x,y))
```

10. **To Generate A Matrix Product Of Two Arrays**

```python
x=[[1,0],[1,1]]
y=[[3,1],[2,2]]
print(np.matmul(x,y))
```

11. **To Find The Roots Of The Following Polynomials**

```python
print(np.roots([1,-2,1]))
print(np.roots([1,-12,10,7,-10]))
```

12. **To Compute The Following Polynomial Values**

np.polyval()->To Evaluate A Polynomial At Specific Values Of x

```python
print(np.polyval([1,-2,1],2))
print(np.polyval([1,-12,10,7,-10],3))
```

13. **To Add One Polynomial To Another,Substract One Polynomial From Another,Multiply One Polynomial With Another And Divide One Polynomial By Another**

```python
from numpy.polynomial import polynomial as P
x=(10,20,30)
y=(30,40,50)
print(P.polyadd(x,y))
print(P.polysub(x,y))
print(P.polymul(x,y))
print(P.polydiv(x,y))
```

14. **To Compute The Trigonometeric sine,cosine and tangent array Of Angles In Degrees**

```python
print("sine:",np.sin(np.array((0.,30.,45.,60.,90.))*np.pi/180.),"\n")
print("cosine:",np.cos(np.array((0.,30.,45.,60.,90.))*np.pi/180.),"\n")
print("tangent:",np.tan(np.array((0.,30.,45.,60.,90.))*np.pi/180.),"\n")
```

15. **To Calculate Inverse sine,cos and tangent For All Elements In A Given Array**

```python
x=np.array([-1.,0,1.])
print("Inverse Sine:",np.arcsin(x),"\n")
print("Inverse Cosine:",np.arccos(x),"\n")
print("Inverse Tangent:",np.arctan(x))
```
