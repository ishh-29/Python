35. **To Add Elements To A Matrix. If An Element In The Matrix Is 0, We'll Not Add The Element Below This Element.**

```python
def summat(m):
  arr=np.array(m)
  sum=0
  for i in range(len(arr)):
    for j in range(len(arr[i])):
      if arr[i][j]==0 and i<len(arr)-1:
        arr[i+1][j]=0
      sum+=arr[i][j]
  return sum
mat=[[1, 1, 0, 2],
     [0, 3, 0, 3],
     [1, 0, 4, 4]]
print(mat)
print("Sum:",summat(mat))
```

38. **To Generate A Uniform,Non-Uniform Random Sample From A Given 1-D Array With And Without
Replacement**

Sampling With Replacement-> After An Item Is Selected,It Is Returned To The Population Before The Next Draw.This Means The Population Size Stays The Same For Each Selection.

Sampling Without Replacement->Once An Item Is Selected,It Is Not Returned To The Population.The Population Size Decreases With Each Draw.

```python
print("Generating A Uniform Random Sample With Replacement:")
print(np.random.choice(7,5))
print("Generating A Uniform Random Sample Without Replacement:")
print(np.random.choice(7,5,replace=False))
print("Generating A Non-Uniform Random Sample With Replacement:")
print(np.random.choice(7,5,p=[0.1,0.2,0,0.2,0.4,0,0.1]))
#p Contains Custom Probabilites
print("Generating A Uniform Random Sample Without Replacement:")
print(np.random.choice(7,5,replace=False,p=[0.1,0.2,0,0.2,0.4,0,0.1]))
```
